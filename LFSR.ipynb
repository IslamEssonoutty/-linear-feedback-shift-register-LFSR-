```python
# Function to convert a string into a stream of binary bits
def String_to_bits(Input_string):
    Binary_string = ""

    for char in Input_string:
        binary_char = bin(ord(char))[2:]
        padded_binary_char = binary_char.zfill(8)
        Binary_string += padded_binary_char
        
    return Binary_string


def xor_gate(a, b):
    if (a or b) and not (a and b):
        return 1
    else:
        return 0


# Function to generate a key using a shift register
def Key_generation(Flip_flop_state):
    Key = ""
    Clk = 0
    Length = 4

    print("Clk  S0 S1 S2 S3")
    print(str(Clk) + " " + str(Flip_flop_state))
    Key = Key + str(Flip_flop_state[0])
    
    while Clk < ((2**Length)-1):
        s0 = Flip_flop_state [0] 
        s1 = Flip_flop_state [1] 
        s2 = Flip_flop_state [2] 
        s3 = Flip_flop_state [3] 
        
        temp = xor_gate(Flip_flop_state [0] , Flip_flop_state [2])
        temp = xor_gate(temp, Flip_flop_state [3])

        Flip_flop_state[3] = temp
        Flip_flop_state[2] = s3
        Flip_flop_state[1] = s2
        Flip_flop_state[0] = s1
        
        Clk = Clk + 1
        print(str(Clk) + " " + str(Flip_flop_state))
        Key = Key + str(Flip_flop_state[0])
    print()
    
    return Key

# Function to encrypt a message using the generated key
def Encrypte(Message, Key):
    Encrypted_message = ""
    Key_position = 0
    
    for bit in Message:
        Char = xor_gate(int(bit), int(Key[Key_position]))
        
        Encrypted_message = Encrypted_message + str(Char)
        Key_position = Key_position + 1
        if Key_position >= 15:
            Key_position = 0
        
    return Encrypted_message

# Function to convert a binary string into characters
def Binary_string_to_characters(Binary_str):
    Characters = ""
    Length = len(Binary_str) // 8  # Divide by 8 to get the number of characters
    Counter = 0
    
    while Counter < Length:
        Byte = Binary_str[:8]  # Get the first 8 bits
        Binary_str = Binary_str[8:]  # Remove the first 8 bits
        Char = int(Byte, 2)  # Convert the binary representation to an integer
        Char = chr(Char)  # Convert the integer to a character
        Characters = Characters + str(Char)
        Counter = Counter + 1
    
    return Characters


def Decrypte(Message, Key):
    Decrypted_message = ""
    Key_position = 0
    
    for bit in Message:
        Char = xor_gate(int(bit), int(Key[Key_position]))
        
        Decrypted_message = Decrypted_message + str(Char)
        Key_position = Key_position + 1
        if Key_position >= 15:
            Key_position = 0
        
    return Decrypted_message
```
